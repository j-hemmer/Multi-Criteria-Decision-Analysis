# Retrieving the hospital X and Y coordinates

```python
import pandas as pd
import regex
```

```python
df = pd.read_csv('hospitals.csv')
df.head()
```

```python
#POINT (-122.43103755 37.79142444)
#get coordinates from data frame
coordinates = df['point']

#Function that extracts x and y coordinates. Returns none if there is an issue
def clean(data):
    data = str(data)
    matches = regex.findall(r'\((\-?\d{1,5}\.\d{1,})\s(\d{1,5}\.\d{1,})', data)
    return matches[0] if matches else (None, None)

#Adds the x and y coordinates to the dataframe
df[['x', 'y']] = pd.DataFrame(coordinates.apply(clean).tolist(), columns=['x', 'y'])

#removes all the other rows besides x and y and the hopsital name
df = df[['Facility Name', 'x', 'y', 'Services']]
df = df[df['Services'] == 'Hospital']
#creates a new csv
df.to_csv('hospitals_coords.csv', index=False)
```
